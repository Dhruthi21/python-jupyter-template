### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.arange(1, 6)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Error handling

# Check if x and y have the same length
if len(x) != len(y):
    raise ValueError("x and y must have the same length for the line plot.")

# Check if categories and values have the same length for the bar plot
if len(categories) != len(values):
    raise ValueError("The categories and values arrays must have the same length for the bar plot.")

# Check if scatter_x and scatter_y have the same length for the scatter plot
if len(scatter_x) != len(scatter_y):
    raise ValueError("scatter_x and scatter_y must have the same length for the scatter plot.")

# Check if pie_labels and pie_sizes have the same length for the pie chart
if len(pie_labels) != len(pie_sizes):
    raise ValueError("The pie_labels and pie_sizes arrays must have the same length for the pie chart.")

# Check if pie_sizes contains non-negative values and is not empty
if any(size < 0 for size in pie_sizes):
    raise ValueError("Pie chart sizes must be non-negative.")
if sum(pie_sizes) == 0:
    raise ValueError("The sum of pie_sizes must be greater than 0.")

# Check if all values in the bar chart are numeric
if not all(isinstance(val, (int, float)) for val in values):
    raise TypeError("The values array for the bar chart must contain numeric values.")

# Check if categories, labels, or data are empty
if not categories or not values or not scatter_x or not scatter_y or not pie_labels or not pie_sizes:
    raise ValueError("None of the input arrays (categories, values, scatter_x, scatter_y, pie_labels, pie_sizes) can be empty.")

# Create a figure with a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot (Top-Left)
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('X values')
axes[0, 0].set_ylabel('Y values')

# Bar Chart (Top-Right)
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')

# Scatter Plot (Bottom-Left)
axes[1, 0].scatter(scatter_x, scatter_y, color='blue', marker='o')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X values')
axes[1, 0].set_ylabel('Y values')

# Pie Chart (Bottom-Right)
axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'orange'])
axes[1, 1].set_title('Pie Chart')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()

```
